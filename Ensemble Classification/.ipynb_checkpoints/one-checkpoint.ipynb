```python
import numpy as np
import pandas as pd

# Plotting Libs
import seaborn as sns

# Pipe Creation
from sklearn import set_config
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin

# Data based Imports
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, FunctionTransformer

# Classifiers
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

# Ensemble Module
from sklearn.ensemble import VotingClassifier

# Metrics
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# Data Load

```python
X, y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=420)
```

# Defining Different Classifiers

```python
lr = LogisticRegression(max_iter=200, random_state=420)
dt_cl = DecisionTreeClassifier(max_depth=4, random_state=420)
svc = SVC(probability=True, random_state=420)
```

```python
voting_clf = VotingClassifier(
    estimators=[
        ('lr', lr),
        ('dt', dt_cl),
        ('svc', svc)
    ],
    voting='soft'
)
```

```python
voting_clf.fit(X_train, y_train)
```

```python
y_pred, y_pred_pro = voting_clf.predict(X_test), voting_clf.predict_proba(X_test)
```

```python
arr = np.hstack(
    (
        y_test.reshape(30, 1), y_pred.reshape(30, 1), y_pred_pro
    )
)
```

```python
test_results = pd.DataFrame(arr, columns=["y_test", "y_pred", "y_pred_pro_0", "y_pred_pro_1", "y_pred_pro_2"])
```

```python
test_results[test_results.y_test != test_results.y_pred]
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, cmap='viridis', annot=True, alpha=0.5, linewidths=1, linecolor='k')
```

```python
print(classification_report(y_test, y_pred))
```

# Pipeline in Sklearn

```python
pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(max_iter=200, random_state=420))
])
pipe_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(probability=True, random_state=420))
])
```

```python
voting_clf = VotingClassifier(
    estimators=[
        ('Logistic Pipe', pipe_lr),
        ('Decision Tree', dt_cl),
        ('SVC Pipe', pipe_svc)
    ], voting='soft'
)
```

```python
voting_clf.fit(X_train, y_train)
```

```python
y_pred, y_pred_pro = voting_clf.predict(X_test), voting_clf.predict_proba(X_test)
```

```python
arr = np.hstack(
    (
        y_test.reshape(30, 1), y_pred.reshape(30, 1), y_pred_pro
    )
)
```

```python
test_results = pd.DataFrame(arr, columns=["y_test", "y_pred", "y_pred_pro_0", "y_pred_pro_1", "y_pred_pro_2"])
```

```python
test_results[test_results.y_test != test_results.y_pred]
```

# Custom Transformation

```python
# Function based transformer
def log_first_feature(X, n=None):
    X = X.copy()
    if n is not None:
        X[:, 0] = np.log(X[:, 0]) / np.log(n)
    else:
        X[:, 0] = np.log1p(X[:, 0])
    return X
```

```python
# Class Based Transformation
class AddSquaredFeatures(BaseEstimator, TransformerMixin):
    """Adding the squared features as mew column in the data"""
    def __init__(self, active=True):
        self.active = active

    def fit(self, X, *args, **kwargs):
        return self

    def transform(self, X, y=None, *args, **kwargs):
        X = X.copy()
        if self.active:
            squared_features = X[:, [0]] ** 2
            return np.hstack([X, squared_features])

        return X
```

```python
# Using lambda as a generator function ...
log_transformer_10 = FunctionTransformer(lambda x: log_first_feature(x, 10))
log_transformer = FunctionTransformer(lambda x: log_first_feature(x, n=None))
```

```python
pipe_lr_lg_trans = Pipeline([
    ('log_transform', log_transformer),
    ('scale', StandardScaler()),
    ('lr', LogisticRegression(max_iter=200, random_state=1))
])
```

```python
pipe_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(probability=True, random_state=420))
])
```

```python
pipe_dt_cl = Pipeline([
    ('feature-engineering', AddSquaredFeatures()),
    ('decision-tree', DecisionTreeClassifier(max_depth=4, random_state=420)),
])

pipe_dt_cl
```

```python
voting_custom_pipe = VotingClassifier(
    estimators=[
        ('lr_transformation', pipe_lr_lg_trans),
        ('svc', pipe_svc),
        ('decision-tree-pipe', pipe_dt_cl)
    ], voting='soft'
)

voting_custom_pipe
```

```python
models = {
    'lr-trans_pipe': pipe_lr_lg_trans,
    'svc-pipe': pipe_svc,
    'decision-tree-pipe': pipe_dt_cl,
    'ensemble-model-pipe': voting_custom_pipe
}

results = {}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_train)
    results[name] = accuracy_score(y_train, y_pred)
```
